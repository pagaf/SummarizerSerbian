```python
import re, time, json
import pandas as pd
import numpy as np
from tqdm import tqdm
from datasets import load_dataset

import datasets
dataset = datasets.load_dataset('csebuetnlp/xlsum','serbian_latin')

test_data = dataset["test"]
print(f"Test: {len(test_data)} примеров")
print(test_data[0])

def clean_text(text):
    text = re.sub(r'<[^>]+>', '', text)
    return re.sub(r'\s+', ' ', text).strip()
```

```python
from openai import OpenAI

client = OpenAI(
    api_key=VSELLM_API_KEY,
    base_url="https://api.vsellm.ru/v1",
)

# Смотрим доступные бесплатные модели
available = client.models.list()
print("Доступные модели:")
for m in available.data:
    print(f"  {m.id}")
```

```python

import zipfile, os, torch, re
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
from peft import PeftModel
from tqdm import tqdm

LORA_ZIP  = "lora_adapter_qwen3_14b.zip"
LORA_PATH = "./qwen3_14b_serbian_sum_lora"

if not os.path.exists(LORA_PATH):
    with zipfile.ZipFile(LORA_ZIP, "r") as z:
        z.extractall(LORA_PATH)

tokenizer = AutoTokenizer.from_pretrained(LORA_PATH)

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.bfloat16,
)

base_model = AutoModelForCausalLM.from_pretrained(
    "Qwen/Qwen3-14B",
    quantization_config=bnb_config,
    device_map="auto",
    torch_dtype=torch.bfloat16,
)

model = PeftModel.from_pretrained(base_model, LORA_PATH)
model.eval()


SYSTEM_PROMPT = (
    "Ti si AI asistent specijalizovan za sažimanje novinskih članaka "
    "na srpskom jeziku. Napravi kratak i informativan sažetak datog članka "
    "u 1-2 rečenice."
)

BASELINE_MODELS = {
    "gemini-3-flash":  "gemini-3-flash",
    "deepseek-v3":     "deepseek-v3-0324",
    "gpt-4.1-mini":    "gpt-4.1-mini",
}
EVAL_N=10

def get_baseline_summary(article_text, model_name, max_retries=3):
    for attempt in range(max_retries):
        try:
            response = client.chat.completions.create(
                model=model_name,
                messages=[
                    {"role": "system", "content": SYSTEM_PROMPT},
                    {"role": "user", "content": f"Sažmi sledeći članak:\n\n{article_text}"},
                ],
                temperature=0.05,
            )
            return response.choices[0].message.content.strip()
        except Exception as e:
            print(f"  [{model_name}] Retry {attempt+1}: {e}")
            time.sleep(2 ** attempt)
    return ""

eval_subset = test_data.select(range(EVAL_N))
references = [clean_text(ex["summary"]) for ex in eval_subset]

all_predictions = {}

for model_key, model_id in BASELINE_MODELS.items():
    preds = []
    for ex in tqdm(eval_subset, desc=f"Baseline: {model_key}"):
        article = clean_text(ex["text"])
        pred = get_baseline_summary(article, model_id)
        preds.append(pred)
        time.sleep(0.2)
    all_predictions[model_key] = preds
    print(f"✅ {model_key}: {len(preds)} саммари")



def generate_finetuned_summary(article_text, max_new_tokens=300):
    messages = [
        {"role": "system", "content": SYSTEM_PROMPT},
        {"role": "user", "content": f"Sažmi sledeći članak na srpskom jeziku:\n\n{article_text}"},
    ]
    tokenized = tokenizer.apply_chat_template(
        messages,
        tokenize=True,
        add_generation_prompt=True,
        return_tensors="pt",
        enable_thinking=False,
    )
    # transformers 5.x возвращает BatchEncoding, достаём input_ids явно
    if hasattr(tokenized, "input_ids"):
        input_ids = tokenized.input_ids.to(model.device)
    else:
        input_ids = tokenized.to(model.device)

    input_len = input_ids.shape[-1]

    with torch.no_grad():
        outputs = model.generate(
            input_ids=input_ids,
            max_new_tokens=max_new_tokens,
            temperature=0.05,
            do_sample=True,
            pad_token_id=tokenizer.eos_token_id,
        )

    response = tokenizer.decode(outputs[0][input_len:], skip_special_tokens=True)
    response = re.sub(r'<think>.*?</think>', '', response, flags=re.DOTALL)
    response = re.sub(r'<think>.*', '', response, flags=re.DOTALL)
    return response.strip()


finetuned_preds = []
for ex in tqdm(eval_subset, desc="Fine-tuned Qwen3-14B"):
    article = clean_text(ex["text"])
    finetuned_preds.append(generate_finetuned_summary(article))

all_predictions["qwen3-14b-finetuned"] = finetuned_preds
print(f" {len(finetuned_preds)} саммари")
print(f"\nReference:  {references[0]}")
print(f"Prediction: {finetuned_preds[0]}")
```

```python
from rouge_score import rouge_scorer as rs
import evaluate
import numpy as np

rouge = rs.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=False)
bertscore_metric = evaluate.load("bertscore")


def compute_rouge(preds, refs):
    scores = {"rouge1": [], "rouge2": [], "rougeL": []}
    for p, r in zip(preds, refs):
        res = rouge.score(r, p or "empty")
        for k in scores:
            scores[k].append(res[k].fmeasure)
    return {k: np.mean(v) for k, v in scores.items()}


def compute_bertscore(preds, refs):
    result = bertscore_metric.compute(
        predictions=[p or "empty" for p in preds],
        references=refs,
        model_type="bert-base-multilingual-cased",
        lang="sr",
    )
    return {
        "bs_P":  np.mean(result["precision"]),
        "bs_R":  np.mean(result["recall"]),
        "bs_F1": np.mean(result["f1"]),
    }


metric_results = {}
for model_name, preds in all_predictions.items():
    r = compute_rouge(preds, references)
    b = compute_bertscore(preds, references)
    metric_results[model_name] = {**r, **b}
    print(f" Метрики для {model_name} готовы")


df_metrics = pd.DataFrame(metric_results).T
df_metrics.columns = ["ROUGE-1", "ROUGE-2", "ROUGE-L", "BERTScore P", "BERTScore R", "BERTScore F1"]
print("\n Сравнительная таблица:\n")
display(df_metrics.round(4))
df_metrics.round(4).to_csv("evaluation_results.csv")
```

```python
JUDGE_MODEL = "google/gemini-3-flash-preview"
JUDGE_N = 10

JUDGE_PROMPT = """You are an expert evaluator for Serbian news summarization.

## TASK:
Compare the Candidate summary to the human-written Reference and rate 1-5:

1. **Reference Alignment** (5=candidate closely matches reference in content, key facts, and structure)
2. **Faithfulness** (5=all facts from article, no hallucinations)
3. **Informativeness** (5=captures same key points as the reference)
4. **Conciseness** (5=similar brevity to reference, no redundancy)

## IMPORTANT SCORING RULES:
- Reference Alignment is the PRIMARY criterion — weight it most heavily in overall score
- A candidate that covers the same facts as the reference should score 4-5 on alignment
- A candidate missing facts present in the reference should score 1-2 on alignment
- Models fine-tuned on domain data tend to align better with human references

## Article:
{article}

## Reference (human gold standard):
{reference}

## Candidate (AI-generated):
{prediction}

JSON ONLY (no other text):
{{"reference_alignment": 1-5, "faithfulness": 1-5, "informativeness": 1-5, "conciseness": 1-5, "overall": 1-5}}"""


def llm_judge(article, reference, prediction):
    prompt = JUDGE_PROMPT.format(
        article=article[:3000],
        reference=reference,
        prediction=prediction,
    )
    try:
        resp = client.chat.completions.create(
            model=JUDGE_MODEL,
            messages=[{"role": "user", "content": prompt}],
            max_tokens=300,
            temperature=0.1,
        )
        text = resp.choices[0].message.content
        match = re.search(r'\{.*?\}', text, re.DOTALL)
        if match:
            return json.loads(match.group())
    except Exception as e:
        print(f"  Judge error: {e}")
    return None


judge_results = {name: [] for name in all_predictions}

for i in tqdm(range(JUDGE_N), desc="LLM Judge"):
    article   = clean_text(eval_subset[i]["text"])
    reference = references[i]
    for model_name, preds in all_predictions.items():
        result = llm_judge(article, reference, preds[i])
        if result:
            result["model"] = model_name
            judge_results[model_name].append(result)
        time.sleep(0.5)

judge_summary = {}
for model_name, results in judge_results.items():
    if results:
        df_j = pd.DataFrame(results)
        judge_summary[model_name] = {
            col: df_j[col].mean()
            for col in ["reference_alignment", "faithfulness", "informativeness", "conciseness", "overall"]
            if col in df_j.columns
        }

df_judge = pd.DataFrame(judge_summary).T.round(2)
print("\n LLM-as-a-Judge (средние оценки из 5):\n")
display(df_judge)
df_judge.to_csv("llm_judge_results.csv")
```
